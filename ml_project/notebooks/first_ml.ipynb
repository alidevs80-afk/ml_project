```python
# Import required libraries
from sklearn.datasets import fetch_california_housing
import pandas as pd


def load_and_explore_california_housing():
    try:
        # Load the dataset
        housing = fetch_california_housing(
            as_frame=True
        )  # as_frame=True returns a pandas DataFrame

        # Extract features and target
        X = housing.data  # Features DataFrame
        y = housing.target  # Target Series (median house value in $100,000s)

        # Combine into one DataFrame for easy viewing
        df = pd.concat([X, y.rename("MedHouseValue")], axis=1)

        df = df.to_csv("california_housing.csv", index=False)  # Save to CSV file

        # # Display dataset info
        # print("=== California Housing Dataset ===")
        # print(f"Shape: {df.shape}")
        # print("\nFeature Names:", list(X.columns))
        # print("\nFirst 5 rows:")
        # print(df.head())

        # # Basic statistics
        # print("\nSummary Statistics:")
        # print(df.describe())

        # return df

    except Exception as e:
        print(f"Error loading dataset: {e}")
        return None


if __name__ == "__main__":
    df = load_and_explore_california_housing()
```

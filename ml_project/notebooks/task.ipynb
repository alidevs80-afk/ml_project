```python
import pandas as pd 
import numpy as np 
data = { 
'Nitrogen_Level': [20.5, 21.3, 19.8, 22.1, 20.0, 23.5, 24.1, np.nan, 23.8, 19.5], 
'Phosphorus_Level': [45.1, 46.2, 44.0, 48.5, 45.0, 55.2, np.nan, 56.1, 54.8, 43.5], 
'Rainfall_mm': [105, 110, 102, 115, 108, 150, 155, 160, 152, 100], 
'Crop_Type': ['Wheat', 'Wheat', 'Wheat', 'Wheat', 'Wheat', 'Rice', 'Rice', 'Rice', 'Rice', 'Wheat'] 
} 
df_soil = pd.DataFrame(data)
```

```python
df_soil
```

```python
df_soil.isnull().sum()
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=2, weights='distance')
df_soil[['Nitrogen_Level', 'Phosphorus_Level']] = imputer.fit_transform(df_soil[['Nitrogen_Level', 'Phosphorus_Level']])
```

```python
df_soil.isnull().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
features = df_soil.drop('Crop_Type', axis=1)
labels = df_soil['Crop_Type']

scaler = StandardScaler()
soil_scaled_features = scaler.fit_transform(features)  
```

```python

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(soil_scaled_features, labels)

new_soil = pd.DataFrame({'Nitrogen': [8], 'Phosphorus': [3], 'Rainfall': [145]})

prediction = knn.predict(new_soil)

print(f"Features of the new soil sample:\n{new_soil}\n")
print(f"--> The KNN model predicts this soil will grow '{prediction[0]}'.")
```

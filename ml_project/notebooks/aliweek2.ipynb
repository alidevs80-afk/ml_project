```python
# What is K-Nearest Neighbors (KNN)?
# K-Nearest Neighbors (KNN) is one of the simplest and most intuitive machine learning algorithms. It is primarily used for classification (grouping things into categories).

# Think of it like moving to a new neighborhood. If you want to know what kind of neighborhood it is, you look at your closest neighbors. If the majority of your 5 closest neighbors are students, you can guess it's a student neighborhood.

# How the algorithm works:

# You have a dataset of points that already have labels (e.g., Class 0 and Class 1).

# You introduce a new, unknown point.

# The algorithm calculates the distance (usually straight-line/Euclidean distance) from the new point to all the existing points in the dataset.

# It finds the "K" closest points (where K is a number you choose, like 3 or 5).

# The new point is assigned the class that is most common among those K neighbors (a simple majority vote).
```

```python
# Cell 1: Import necessary libraries
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Cell 2: Create the Seeds DataFrame
# Creating synthetic data for 3 different seed families
data = {
    'Length_mm': [4.5, 4.7, 4.1, 5.8, 6.0, 5.9, 3.1, 3.3, 3.0],
    'Weight_g': [2.1, 2.3, 2.0, 3.5, 3.8, 3.6, 1.1, 1.2, 1.0],
    'Family': ['Wheat', 'Wheat', 'Wheat', 'Corn', 'Corn', 'Corn', 'Rice', 'Rice', 'Rice']
}

# Convert the dictionary into a Pandas DataFrame
df_seeds = pd.DataFrame(data)

# Display the DataFrame to verify it looks correct
df_seeds
```

```python
# Cell 3: Apply KNN and predict the family of a new upcoming seed

# 1. Separate the features (X) from the target label (y)
X = df_seeds[['Length_mm', 'Weight_g']]
y = df_seeds['Family']

# 2. Initialize the KNN Classifier 
# We'll set K=3 to look at the 3 nearest neighbors
knn = KNeighborsClassifier(n_neighbors=3)

# 3. Train the model using our DataFrame
knn.fit(X, y)

# 4. Define our new, upcoming seed as a DataFrame (to match feature names)
# Let's say we found a seed that is 4.6mm long and weighs 2.2g
new_seed = pd.DataFrame({'Length_mm': [8], 'Weight_g': [3]})

# 5. Predict the family of this new seed
prediction = knn.predict(new_seed)

print(f"Features of the new seed:\n{new_seed}\n")
print(f"--> The KNN model predicts this seed belongs to the '{prediction[0]}' family.")
```

```python
# What is KNN Imputation?
# KNN Imputation (K-Nearest Neighbors Imputation) is a technique used to fill in missing data points by borrowing information from the most similar rows in your dataset.

# Instead of taking a lazy approach—like filling every missing value with the overall dataset average or deleting the row entirely—KNN imputation makes an educated guess. It assumes that rows with similar data in other columns will likely have similar values in the column that is missing.

# Its Role in Preprocessing
# When dealing with the Titanic dataset, you often find that the Age column has a lot of missing values. If you drop every passenger missing an age, you lose valuable data. If you fill all missing ages with the average age (e.g., 29), you distort the data (a 1st-class male passenger traveling alone is likely older than a 3rd-class child traveling with siblings).

# The role of KNN imputation is to preserve the variance and relationships in your data by making a highly contextual, personalized guess for each missing value.

# How it Works (The Titanic Example)
# Let’s say Passenger A is missing their Age.

# Feature Matching: The algorithm looks at Passenger A's other available features (e.g., Fare = $70, Pclass = 1st Class, SibSp = 0).

# Finding the Neighbors: It calculates the distance between Passenger A and every other passenger in the dataset based on those available features. It isolates the K closest matches (where "K" is a number you choose, like 5).

# Imputing the Value: It looks at the known Age of those 5 closest neighbors. If their ages are 42, 45, 39, 44, and 40, the algorithm calculates the average (42).

# Filling the Blank: Passenger A's missing Age is replaced with 42.
```

```python
# Here is the difference in plain English:

# KNN Classifier (What we used previously): Predicts a category. You have a brand-new, completely measured seed, and you want to know what family it belongs to. The neighbors take a majority vote (e.g., "2 of us are Wheat and 1 is Rice, so you must be Wheat").

# KNN Imputer: Predicts a missing number. You have a seed already in your dataset, but your scale broke, so its weight is missing. It looks at the closest seeds and takes the average of their weights to fill in your blank space.
```

```python
# Cell 1: Setup and Complex Data Generation
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# 1. Simulate 100 seeds with 7 complex features
np.random.seed(42)
n_seeds = 100

data = {
    'Area_mm2': np.random.uniform(10.0, 21.0, n_seeds),
    'Perimeter_mm': np.random.uniform(12.0, 17.0, n_seeds),
    'Compactness': np.random.uniform(0.80, 0.92, n_seeds),
    'Length_mm': np.random.uniform(4.8, 6.7, n_seeds),
    'Width_mm': np.random.uniform(2.5, 4.1, n_seeds),
    'Asymmetry_Coef': np.random.uniform(1.0, 8.0, n_seeds),
    'Groove_Length_mm': np.random.uniform(4.5, 6.5, n_seeds),
    'Family': np.random.choice(['Wheat', 'Corn', 'Rice'], n_seeds)
}

df_seeds_production = pd.DataFrame(data)

# 2. Simulate "Sensor Failures" by randomly dropping 15% of the data in specific columns
for col in ['Weight_g', 'Compactness', 'Asymmetry_Coef']:
    # Let's pretend our "Weight" sensor is actually derived from Area for this simulation
    if col == 'Weight_g':
        df_seeds_production[col] = df_seeds_production['Area_mm2'] * 0.15 + np.random.normal(0, 0.1, n_seeds)
        
    mask = np.random.rand(n_seeds) < 0.15 # 15% chance of failure
    df_seeds_production.loc[mask, col] = np.nan

print("Dataset Generated. Shape:", df_seeds_production.shape)
df_seeds_production.head(10)
```

```python
# New Cell: Deep Dive into Missing Data and Duplicates

# 1. Check for Duplicate Rows
duplicate_count = df_seeds_production.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")

if duplicate_count > 0:
    print("\nHere are the duplicate rows:")
    display(df_seeds_production[df_seeds_production.duplicated(keep=False)])
else:
    print("\nNo duplicate rows found. Proceeding to missing values check...")

print("\n" + "="*50 + "\n")

# 2. Detailed Breakdown of Missing Values
print("Total missing values per column:")
print(df_seeds_production.isnull().sum())

print("\n--- Let's look at the specific rows that have missing data ---")

# This filter grabs ANY row that has at least one NaN in ANY column
rows_with_missing_data = df_seeds_production[df_seeds_production.isnull().any(axis=1)]

print(f"Total rows affected by sensor failures: {len(rows_with_missing_data)} out of 100")

# Using 'display' instead of 'print' renders a clean HTML table in Jupyter
display(rows_with_missing_data)
```

```python
# Cell 3: Feature Scaling (Must run before Cell 4!)

# 1. Separate the text labels (Family) from the numbers
features = df_seeds_production.drop('Family', axis=1)
labels = df_seeds_production['Family']

# 2. Initialize the StandardScaler
scaler = StandardScaler()

# 3. Scale the features
# Note: StandardScaler ignores NaNs automatically during the calculation
scaled_features = scaler.fit_transform(features)

# 4. Convert back to DataFrame and create the variable 'df_scaled'
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)
print("Data successfully scaled. 'df_scaled' is now ready for imputation.")
```
